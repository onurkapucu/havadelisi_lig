```python
def check_sicaklik_faul(tahmin1,tahmin2):
    faul = abs(tahmin1 - tahmin2) < 0.2
    return faul
```

```python
def check_sicaklik_offside(tahmin1,tahmin2,gerceklesen):
    offside = min(abs(tahmin1-gerceklesen),abs(tahmin2-gerceklesen)) > 3
    return offside
```

```python
def sicaklik_gol(tahmin1,tahmin2,gerceklesen):
    faul = check_sicaklik_faul(tahmin1,tahmin2)
    offside = check_sicaklik_offside(tahmin1,tahmin2,gerceklesen)
    tahmin1_diff = abs(tahmin1-gerceklesen)
    tahmin2_diff = abs(tahmin2-gerceklesen)
    
    if tahmin1_diff < tahmin2_diff:
        tahminci = tahmin1
        oran = tahmin1_diff/tahmin2_diff
    else:
        tahminci = tahmin2
        oran = tahmin2_diff/tahmin1_diff
        
    if (faul==0) & (offside==0):
        gol = oran < oran_sic
    else:
        gol = False
    
    return gol,tahminci
```
